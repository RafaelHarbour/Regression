# Looking into the relationship of energy prices increase and the effect it has on different CPI divisions

### First we load the data, each dataset includes the CPI of the relevant division and the current price of each energy source

```python
import pandas as pd
dataset_clothing = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Clothing and footwear")
dataset_alcohol = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Alcoholic beverages and tobacco")
dataset_education = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Education")
dataset_misc = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Miscellaneous goods and service")
dataset_communication = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Communication")
dataset_furniture = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Furniture and furnishings")
dataset_health = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Health")
dataset_culture = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Recreation and culture")
dataset_food = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Food and non-alcoholic beverage")
dataset_hotels = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Restaurants and hotels")
dataset_transport = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Transport")
dataset_housing = pd.read_excel("datasets/Energy-CPI.xlsx", sheet_name="Housing and household services")
r2_values =  []
```

# Clothing and footwear

```python
corr_matrix = dataset_clothing.corr()
corr_matrix["CPI"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix
attributes = ["CPI", "Electricity", "Gas",
"Liquid Fuels", "Solid Fuels", "Motor fuel and oil" ]
scatter_matrix(dataset_clothing[attributes], figsize=(16, 12))
```

```python
y = dataset_clothing["CPI"]
X = dataset_clothing.drop(columns = ["CPI"])
```

```python
import statsmodels.api as sm
X= sm.add_constant(X)
mod = sm.OLS(y, X)
res = mod.fit()
print(res.summary())
r2_values.append(res.rsquared)
```
